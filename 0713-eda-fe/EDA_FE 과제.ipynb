### 5) Feature Engineering

feature에 대한 전처리를 하기 이전에 결측치가 있는 column을 모두 제거했는데, 다시 초기 상태로 되돌리고 진행하겠습니다.

```python
df = pd.read_csv('airbnb_listings.csv', index_col=0)
```

- #### <font color='orange'> Q12. 결측치에 대한 처리를 하고 싶습니다.
#### <font color='orange'> 주어진 모든 column들에 대하여 결측치가 25%이상인 경우에는 column을 drop하고 10%이상 25%미만인 경우에는 결측치가 있는 row만 drop하고, 10%미만인 경우에는 중앙값으로 대체해주세요.

```python
#Q12
missing_rate = df.isnull().sum() / len(df)
for col in missing_rate.index:
    if missing_rate[col] >= 0.25:
        df.drop(col, axis=1, inplace=True)
    elif missing_rate[col] >= 0.10:
        df.dropna(subset=[col], inplace=True)
    else:
        val = df[col].mode() if df[col].dtype == object else df[col].median()
        df[col].fillna(val, inplace=True)
```

- #### <font color='orange'> Q13. 앞서 진행한 EDA를 바탕으로 FE를 진행하고자 합니다..
#### <font color='orange'> 총 3개의 FE를 진행하시고, 해당 FE를 진행한 이유에 대해 **간단히** 작성해주세요.


```python
df.head()
```

```python
#Q13-1) One-hot Encoding: `neighborhood`는 5개의 범주 중 하나이기 때문
pd.get_dummies(df['neighborhood'])
```

```python
#Q13-2) 지표 변수: `extra_people`에 대한 추가 비용이 0인지 여부
charge_extra_people = []
for charge in df['extra_people']:
    charge = float(charge[1:])
    if charge == 0:
        charge_extra_people.append('f')
    else:
        charge_extra_people.append('t')
df['charge_extra_people'] = charge_extra_people
df.head()
```

```python
#Q13-3) Scaling: review 항목 중 하나는 100점 만점, 나머지는 10점 만점이기 때문에 0~10 사이 값으로 통일
df['review_scores_rating'] /= 10
```
